```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city = pd.read_csv(city_data_to_load)
ride = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
ride_city = pd.merge(ride, city, how="left", on= "city")
# Display the data table for preview
ride_city.head()
```

## Bubble Plot of Ride Sharing Data

```python
urban_cities = ride_city[ride_city["type"] == "Urban"]
suburban_urban_cities = ride_city[ride_city["type"] == "Suburban"]
rural_cities = ride_city[ride_city["type"] == "Rural"]
```

```python
# Obtain the x and y coordinates for each of the three city types
x1 = urban_cities.groupby(["city"])["fare"].count()
y1 = urban_cities.groupby(["city"])["fare"].mean()
z1 = urban_cities.groupby(["city"])["driver_count"].mean()

x2 = suburban_urban_cities.groupby(["city"])["fare"].count()
y2 = suburban_urban_cities.groupby(["city"])["fare"].mean()
z2 = suburban_urban_cities.groupby(["city"])["driver_count"].mean()

x3 = rural_cities.groupby(["city"])["fare"].count()
y3 = rural_cities.groupby(["city"])["fare"].mean()
z3 = rural_cities.groupby(["city"])["driver_count"].mean()

plt.scatter(x1, y1, s=10*z1, c="coral", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Urban")
plt.scatter(x2, y2, s=10*z2, c="skyblue", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Suburban")
plt.scatter(x3, y3, s=10*z3, c="gold", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Rural")

# Incorporate the other graph properties
plt.grid()
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fares ($)")

# Create a legend
lgnd = plt.legend(title='City Types')
for handler in lgnd.legendHandles:
    handler._sizes = [30]
    
# Incorporate a text label regarding circle size
plt.text(42, 35, "Note: \nCircle size correlates with driver count per city.")

# Save Figure
plt.savefig("../Images/pyber_bubble_plot.png")
```
